### Dynamic Promp Template

```python
import os
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("gpt-4.1")
```

```python
from langchain_core.prompts import ChatPromptTemplate

template = ChatPromptTemplate.from_messages([
    ("system", "You are a Competitive Exam Coach like SSC, UPSC, Bank Exams in India. You will understand the doubt from the students and clear their doubt on subject {subject}. Based on the user age ({age}) and level of experience ({experience_level}) in {subject}, try to make them understand the answers. Try to keep the answers short and to the point. Do NOT include any suggestion at the end."),
    ("user", "{doubt}")
])

prompt = template.invoke({
    "subject": "History",
    "age": 20,
    "experience_level": "Intermediate",
    "doubt": "How many times India was invaded during 18th Century?"
})
```

```python
response = model.invoke(prompt)
print(response.content)
```
